```python
input = data.drop(columns = "DEATH_EVENT")
output = data["DEATH_EVENT"]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(input,output,test_size = 0.3)
x_train.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
```

```python
model.fit(x_train,y_train)
value = model.predict(x_test)
value
```

```python
from sklearn.metrics import accuracy_score
acc = accuracy_score(value,y_test)
print("Prediction accuracy: {:.2f}%".format(acc * 100))
```

```python
data["prediction"] = model.predict(input)
data.sample(10)
```
